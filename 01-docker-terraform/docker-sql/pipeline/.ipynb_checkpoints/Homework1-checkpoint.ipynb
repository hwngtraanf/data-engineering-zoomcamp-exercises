```python
import pandas as pd
import datetime
```

```python
df_taxi_zone_lk = pd.read_csv('./taxi_zone_lookup.csv')
```

```python
df_taxi_zone_lk.head()
```

```python
df_greentrip = pd.read_parquet('./green_tripdata_2025-11.parquet')
```

```python
# Question 3
date_upper = datetime.date(2025, 12, 1)
date_lower = datetime.date(2025, 11, 1)

cnt_short_trip_df = df_greentrip.loc[
    (df_greentrip['trip_distance'] <= 1) 
    & (df_greentrip['lpep_pickup_datetime'] >= pd.Timestamp(date_lower)) 
    & (df_greentrip['lpep_pickup_datetime'] < pd.Timestamp(date_upper))
]
len(cnt_short_trip_df)
# 8007
```

```python
# Question 4
longest_trip_distance = df_greentrip.loc[
    (df_greentrip['trip_distance'] < 100) 
    & (df_greentrip['trip_distance'] == max(longest_trip_distance['trip_distance']))
]
longest_trip_distance['lpep_pickup_datetime']
# 2025-11-14 15:36:27
```

```python
# Question 5
pickup_date = datetime.date(2025, 11, 18)

biggest_pickup = df_greentrip.loc[
    df_greentrip['lpep_pickup_datetime'].dt.date == pickup_date
]

top1 = (biggest_pickup.groupby("PULocationID", as_index=False)["total_amount"]
          .sum()
          .sort_values("total_amount", ascending=False)
          .head(1))

location_id = top1['PULocationID'].values[0]

biggest_pickup_zone = df_taxi_zone_lk.loc[
    df_taxi_zone_lk['LocationID'] == location_id
]

biggest_pickup_zone
# 74	Manhattan	East Harlem North	Boro Zone
```

```python
# Question 6
date_upper = datetime.date(2025, 11, 30)
date_lower = datetime.date(2025, 11, 1)

location_df = df_taxi_zone_lk.loc[
    df_taxi_zone_lk['Zone'] == 'East Harlem North'
]

location_id = location_df['LocationID'].values[0]

tip_df = df_greentrip.loc[
    (df_greentrip['PULocationID'] == location_id) 
    & (df_greentrip['lpep_pickup_datetime'] >= pd.Timestamp(date_lower)) 
    & (df_greentrip['lpep_pickup_datetime'] <= pd.Timestamp(date_upper))
]

largest_tip = tip_df.loc[
    (tip_df['tip_amount'] == max(tip_df['tip_amount'])) 
]
do_location_id = largest_tip['DOLocationID'].values[0]

largest_tip_dropoff_zone = df_taxi_zone_lk.loc[
    df_taxi_zone_lk['LocationID'] == do_location_id
]

largest_tip_dropoff_zone
# 138	Queens	LaGuardia Airport	Airports
```
