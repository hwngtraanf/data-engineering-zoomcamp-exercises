```python
from pyspark.sql import functions as F
from datetime import date
```

```python
df \
    .withColumn("pickup_datetime", F.to_date(df.tpep_pickup_datetime)) \
    .filter(F.col("pickup_datetime") == date(2025, 11, 15)) \
    .count()
```

```python
df.head(1)
```

```python
max_trip_duration = df \
    .withColumn("trip_duration", (F.col("tpep_dropoff_datetime") - F.col("tpep_pickup_datetime"))) \
    .select(F.max(F.col("trip_duration"))) \
    .show(truncate=False)
```

```python
!wget https://d37ci6vzurychx.cloudfront.net/misc/taxi_zone_lookup.csv
```

```python
df_zone = spark.read \
    .option("header", "true") \
    .csv('taxi_zone_lookup.csv')
```

```python
df_zone.show()
```

```python
df \
    .groupBy(F.col("PULocationID")) \
    .count() \
    .orderBy(F.col("count").asc()) \
    .limit(1) \
    .join(df_zone, F.col("PULocationID") == F.col("LocationID"), "inner") \
    .select(F.col("LocationID"), F.col("Borough"), F.col("Zone"), F.col("service_zone")) \
    .show(truncate=False)
```
